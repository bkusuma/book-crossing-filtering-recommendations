```python
isbn_review_counts = ratings["ISBN"].value_counts()
isbn_review_10 = isbn_review_counts[isbn_review_counts > 10] # filter for ISBNs with more than 10 reviews
isbn_review_10
```

```python
ratings = ratings[ratings["ISBN"].isin(isbn_review_10.index)]
ratings
```

```python
ratings.duplicated(subset=["User-ID", "ISBN"]).sum()
```
